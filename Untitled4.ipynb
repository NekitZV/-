```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill
```

```python
data = pd.read_csv('wc_matches.csv')
data
```

```python
df = pd.read_excel('football.xlsx')
excel_file = openpyxl.open('football.xlsx', read_only=False)
sheet = excel_file.active
```

```python
rows = 2

for coef_win,coef_draw,coef_fall,score1,score2 in zip(data['prob1'],data['prob2'],data['probtie'],data['score1'],data['score2']):
    sheet[f'B{rows}'] = int(round(coef_win*100,0))
    sheet[f'C{rows}'] = int(round(coef_draw*100,0))
    sheet[f'D{rows}'] = int(round(coef_fall*100,0))
    
    sorted_coef = sorted([coef_win,coef_draw,coef_fall])
    if (score1 > score2) and (coef_win == sorted_coef[2]):
        greenFill = PatternFill(start_color='00D678', end_color='00D678', fill_type='solid')
        sheet[f'B{rows}'].fill = greenFill

    elif (score1 == score2) and (coef_draw == sorted_coef[2]):
        greenFill = PatternFill(start_color='00D678', end_color='00D678', fill_type='solid')
        sheet[f'C{rows}'].fill = greenFill

    elif (score1 < score2) and (coef_fall == sorted_coef[2]):
        greenFill = PatternFill(start_color='00D678', end_color='00D678', fill_type='solid')
        sheet[f'D{rows}'].fill = greenFill

    elif (score1 > score2) and (coef_win == sorted_coef[1]):
        yellowFill = PatternFill(start_color='FFFF00', end_color='FFFF00', fill_type='solid')
        sheet[f'B{rows}'].fill = yellowFill

    elif (score1 == score2) and (coef_draw == sorted_coef[1]):
        yellowFill = PatternFill(start_color='FFFF00', end_color='FFFF00', fill_type='solid')
        sheet[f'C{rows}'].fill = yellowFill

    elif (score1 < score2) and (coef_fall == sorted_coef[1]):
        yellowFill = PatternFill(start_color='FFFF00', end_color='FFFF00', fill_type='solid')
        sheet[f'D{rows}'].fill = yellowFill

    elif (score1 > score2) and (coef_win == sorted_coef[0]):
        redFill = PatternFill(start_color='FF0000', end_color='FFFF00', fill_type='solid')
        sheet[f'B{rows}'].fill = redFill

    elif (score1 == score2) and (coef_draw == sorted_coef[0]):
        redFill = PatternFill(start_color='FF0000', end_color='FFFF00', fill_type='solid')
        sheet[f'C{rows}'].fill = redFill

    elif (score1 < score2) and (coef_fall == sorted_coef[0]):
        redFill = PatternFill(start_color='FF0000', end_color='FFFF00', fill_type='solid')
        sheet[f'D{rows}'].fill = redFill
    
    excel_file.save('football.xlsx')
    rows+=1
```
